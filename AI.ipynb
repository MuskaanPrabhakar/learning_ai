```python
test ='''For the article you pasted, that would eliminate most navigation and code garbage automatically.

What I would do

If the goal is:

"Get a cheap summary of any webpage"

Then:

Remove script/style/nav/footer/header.
Remove pre and code.
Extract only p, h1, h2, h3.
Join text.
Send to AI.
Ask AI to summarize.

That will usually reduce tokens much more than removing duplicate words, while keeping the information intact. Removing duplicate words is risky because it damages meaning, whereas removing code blocks and boilerplate often cuts 50–80% of the text with almost no loss in summary quality.'''
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    separator="", chunk_size=50, chunk_overlap=1 #where there is separator there it will end chunk and therefore doesn't depend in chunk_size
)
texts = text_splitter.split_text(test)
texts
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=3)
tests= text_splitter.split_text(test)
tests
```

```python
def chunker(str, chunk_size =100):
  formated_string=''.join(str.split('\n'))
  words = formated_string.split()
  chunks=[' '.join(words[0:1+chunk_size]) for i in range(0,len(words),chunk_size)]
  print(chunks)

chunker(str)
```

```python
!pip install -q langchain-huggingface
!pip install -q langchain-core
```

```python
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings

documents = [
    Document(page_content="""
    LangChain is a framework for building LLM applications.
    It supports RAG, agents, memory, and tools.
    """)
]


splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)

chunks = splitter.split_documents(documents)
print(chunks)

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
val = [chunk.page_content for chunk in chunks]
print(val)
vecti = embeddings.embed_query(
    val[0]
)
print(f"Embedding dimension: {len(vecti)}")
vectors = embeddings.embed_documents(
    val
)

# print(f"Chunks: {len(chunks)}")
print(f"Embedding dimension: {len(vectors[0])}")
```

# task: from a document, send relatable chunks to ai and answer

```python
from langchain_community.document_loaders import PDFPlumberLoader
docu = PDFPlumberLoader("ss.pdf")
```

```python
data=docu.load() #loading data from docu
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
chunks = splitter.split_documents(data) #making chunks
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
model_name = "sentence-transformers/all-mpnet-base-v2"
my = HuggingFaceEmbeddings(
    model_name=model_name,
)
string_chunks=[chunk.page_content for chunk in chunks]
my.embed_documents(string_chunks)
```

```python
#putting embeddings in chroma
vector_db = Chroma(
    embedding_function= my,
    persist_directory='db',
    collection_name="vectorDB"
)
```

```python
vector_db.add_documents(chunks)
```

```python
query= input("enter your query regarding pdf: ")
```

```python
to_ai= vector_db.search(query=query,search_type="similarity")
```

```python
response =client.responses.create(
    model="meta-llama/llama-3.3-70b-instruct:free",
    instructions="You have to use chunks to provide best of information",
    input=f"Answer this {query} based on {to_ai}"
)
```

# CHAT BOT

```python
from google.colab import userdata
client3 = OpenAI(
    base_url ="https://generativelanguage.googleapis.com/v1beta/",
    api_key=userdata.get("GEMINI2")
)
```

```python
system_prompt={"role":"system", "content":"You are helpful agent who helps people to plan their vacation based on budget and location and have attention to details"}
messages=[system_prompt]
user=input("What're we planning? ")
messages.append({"role":"user", "content":user})
```

```python
completion = client3.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=messages,n=2, max_tokens=300)
print(completion.choices[0].message.content)
```

```python
def ai():
  user=input("enter 'quit' to leave or else reply: ")
  if user=='quit':
     return
  messages.append({"role":"user", "content":user})
  completion = client3.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=messages, max_tokens=300, temperature=0)
  return completion.choices[0].message.content
```

```python
while(user!='quit'):
  messages.append({'role': 'assistant', 'content': completion.choices[0].message.content})
  response = ai()
  if response is None:
    break
  print(response)
```
